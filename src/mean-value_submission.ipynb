```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train=pd.read_csv('../input/train.csv')
df_test=pd.read_csv('../input/test.csv')

df_test.head()
```

```python
print('Total number of  houses in training set: {}'.format(len(df_train.index)))
print('Mean value of sale prices for training set: {}'.format(df_train.SalePrice.mean()))
print('Deviation of sale prices for training set: {}'.format(df_train.SalePrice.std()))

print('Number of repeating houses in training set: {}'.format(np.sum(df_train.Id.value_counts()>1)))


plt.hist(df_train.SalePrice,bins=np.arange(min(df_train.SalePrice),max(df_train.SalePrice),10000),density=True)
plt.title('Log-Histogram of sale prices')
plt.xlabel('Prices')
plt.ylabel('Density')

print()
```

```python
df_submission=pd.DataFrame({'ID':df_test.Id,'SalePrice':df_train.SalePrice.mean()})
df_submission.to_csv('../output/mean_train_value_submission.csv', index=False)
```
